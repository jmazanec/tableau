```python
import pandas as pd
```

```python
june2019 = pd.read_csv("201906-citibike-tripdata.csv")
july2019 = pd.read_csv("201907-citibike-tripdata.csv")
aug2019 = pd.read_csv("201908-citibike-tripdata.csv")
june2020 = pd.read_csv("202006-citibike-tripdata.csv")
july2020 = pd.read_csv("202007-citibike-tripdata.csv")
aug2020 = pd.read_csv("202008-citibike-tripdata.csv")

```

```python
from datetime import datetime
june2019['starttime'] = pd.to_datetime(june2019['starttime'])
june2019['day_of_week'] = june2019['starttime'].dt.day_name()
june2019["month"] = "June"
june2019["year"] = "2019"
july2019['starttime'] = pd.to_datetime(july2019['starttime'])
july2019['day_of_week'] = july2019['starttime'].dt.day_name()
july2019["month"] = "July"
july2019["year"] = "2019"
aug2019['starttime'] = pd.to_datetime(aug2019['starttime'])
aug2019['day_of_week'] = aug2019['starttime'].dt.day_name()
aug2019["month"] = "August"
aug2019["year"] = "2019"
june2020['starttime'] = pd.to_datetime(june2020['starttime'])
june2020['day_of_week'] = june2020['starttime'].dt.day_name()
june2020["month"] = "June"
june2020["year"] = "2020"
july2020['starttime'] = pd.to_datetime(july2020['starttime'])
july2020['day_of_week'] = july2020['starttime'].dt.day_name()
july2020["month"] = "July"
july2020["year"] = "2020"
aug2020['starttime'] = pd.to_datetime(aug2020['starttime'])
aug2020['day_of_week'] = aug2020['starttime'].dt.day_name()
aug2020["month"] = "August"
aug2020["year"] = "2020"
```

```python
summerbummer = [june2019,july2019,aug2019,june2020,july2020,aug2020]

sb = pd.concat(summerbummer)

sb.head()
```

```python
sb.to_csv("combined.csv")
```
